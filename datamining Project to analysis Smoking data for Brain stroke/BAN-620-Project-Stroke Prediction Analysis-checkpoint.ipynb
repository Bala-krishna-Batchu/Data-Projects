1)decision tree

```python
from sklearn.tree import DecisionTreeClassifier


DT_model= DecisionTreeClassifier()

# fit the model
DT_model.fit(train_x,train_y)

# model score
predict_train_DT = DT_model.predict(train_x)
predict_test_DT = DT_model.predict(valid_x)

# accuracy score
DT_train_score = DT_model.score(train_x,train_y)
DT_test_score = DT_model.score(valid_x,valid_y)

# f1-score
DT_f1_score = metrics.f1_score(valid_y, predict_test_DT)

print('Accuracy on Train data set',DT_train_score)
print('Accuracy on Test data set',DT_test_score)
print('F1-score on Test set:',DT_f1_score)
print('\n')
print(metrics.classification_report(valid_y, predict_test_DT))

from dmba import plotDecisionTree, classificationSummary, regressionSummary
plotDecisionTree(DT_model, feature_names=train_x.columns,class_names=DT_model.classes_)



```

```python
tree_nodes = DT_model.tree_.node_count
print('nodes:', tree_nodes)
```

```python
# classificationSummary()-used for getting accuray measures, function to create confusion matrix for the training data partition.
print('Training data Partition')
classificationSummary(train_y, DT_model.predict(train_x))

#  confusion matrix for validation data . 
print()
print('Validation data Partition')
classificationSummary(valid_y, DT_model.predict(valid_x))
```

improving decision tree using gridsearch

```python
from sklearn.model_selection import GridSearchCV

#"""
#parameters= {'max_depth':[7,10],
            #'max_features':[4,5,6],
            #'min_samples_split' : [90,100],
            #'min_samples_leaf' : [20,30]}


#DT = RandomForestClassifier()

#DT_model_tune = GridSearchCV(DT, param_grid = parameters, cv=3)
#DT_model_tune.fit(train_x,train_y)
# Improved grid search parameters using adapting grid based 
param_grid = {
    'max_depth': list(range(2, 30)),  
    'min_impurity_decrease': [0,0.0005,0.001], 
    'min_samples_split': list(range(5, 30)),
}
# we can use GridSearchCV() fucntion for various combinations of DecisionTreeClassifier() using parameters. 
#cv=5 is that 5-fold cross-validation is used in this case,
#n_jobs=-1 means that the availalbe computer memory (CPU) used to make calculations faster.
# AGridSearchCV() fucntion for various combinations of DecisionTreeClassifier() improved parameters. 
gS = GridSearchCV(DecisionTreeClassifier(), 
                param_grid, cv=5, n_jobs=-1)
gS.fit(train_x, train_y)

# Display best improved paramenters of classification tree. 
print()
print(f'score:{gS.best_score_:.4f}')
print('Improved_parameters: ', gS.best_params_)
```
